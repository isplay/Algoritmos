```python
import networkx as nx
import EoN
import matplotlib.pyplot as plt
from collections import defaultdict
print('generating graph G')

G_random_regular_graph = nx.random_regular_graph(3,100)

#we'll initially infect those near the middle
initial_infections = [13]

#the spontaneous transitions
H = nx.DiGraph() 
H.add_edge('Inf', 'Rec', rate = 1.0)

#the induced transitions
J = nx.DiGraph() 
J.add_edge(('Inf', 'Sus'), ('Inf', 'Inf'), rate = 2.0)

# Initial status of nodes
IC = defaultdict(lambda:'Sus') #a "dict", but by default the value is \texttt{'Sus'}. 
for node in initial_infections:
    IC[node] = 'Inf'
    
# Return values
return_statuses = ['Inf', 'Sus', 'Rec'] 

print('doing Gillespie simulation')

sim = EoN.Gillespie_simple_contagion(G_random_regular_graph, H, J, IC, return_statuses, tmax=10,

                                     return_full_data=True)
times_random_regular_graph, D_random_regular_graph = sim.summary()

sim.display(time=6, node_size = 50)
plt.show()

#ani=sim.animate(ts_plots=[['Inf'], ['Sus'], ['Rec']], node_size = 10)
#ani.save('random_regular_graph_beta_2_gamma_1.mp4', fps=5)
```
